$$ \int^{3}_{1} x^{2} + 2x +5 $$

```python
import numpy as np

N1 = 100
N2 = 1000
N3 = 10000
N4 = 100000

a = 1 
b = 3
c = 10

def fungsi(x) :
    return x**2 + 2*x + 5

def integral_montecarlo(N, a, b, c):
    nhit = 0
    for i in range(N):
        r1 = np.random.rand()
        r2 = np.random.rand()
        x_i = a + r1 *(b-a)
        y_i = r2 * c
        fx = fungsi(x_i)
        if y_i<fx:
            nhit += 1
    return c* (b-a) * nhit/N

hasil1 = integral_montecarlo(N1, a, b, c)
hasil2 = integral_montecarlo(N2, a, b, c)
hasil3 = integral_montecarlo(N3, a, b, c)
hasil4 = integral_montecarlo(N4, a, b, c)
print(f'Hasil {N1}: {hasil1}')
print(f'Hasil {N2}: {hasil2}')
print(f'Hasil {N3}: {hasil3}')
print(f'Hasil {N4}: {hasil4}')
```

$$ \int^{1}_{3} \int^{2}_{0} xy^{2}+2x^{2}y+2x+3y=5 $$

```python
import numpy as np

N1 = 100
N2 = 1000
N3 = 10000
N4 = 100000

a = 1 
b = 3
c = 0
d = 2
e = 20

def fungsi(x,y) :
    return x*y**2 + 2*(x**2)*y + 2*x + 3*y + 5


def integral_montecarlo(N, a, b, c, d, e):
    nhit = 0
    for i in range(N):
        r1 = np.random.rand()
        r2 = np.random.rand()
        r3 = np.random.rand()
        x_i = a + r1 * (b - a)
        y_i = c + r2 * (d - c)
        xy_i = r3 * e
        fx = fungsi(x_i,y_i)
        if xy_i<fx:
            nhit += 1
    return e * (b-a) *(d-c) * nhit/N

hasil1 = integral_montecarlo(N1, a, b, c, d, e )
hasil2 = integral_montecarlo(N2, a, b, c, d, e )
hasil3 = integral_montecarlo(N3, a, b, c, d, e )
hasil4 = integral_montecarlo(N4, a, b, c, d, e )
print(f'Hasil {N1}: {hasil1}')
print(f'Hasil {N2}: {hasil2}')
print(f'Hasil {N3}: {hasil3}')
print(f'Hasil {N4}: {hasil4}')
```
